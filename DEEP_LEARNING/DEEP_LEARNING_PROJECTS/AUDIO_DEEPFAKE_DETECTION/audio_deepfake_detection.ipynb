```python
import torchaudio
import torch
from torch.utils.data import Dataset
from transformers import Wav2Vec2Processor
```

```python


class ASVspoofDataset(Dataset):
    def __init__(self, audio_dir, protocol_df, processor, max_len=16000*4):
        self.audio_dir = audio_dir
        self.df = protocol_df.reset_index(drop=True)
        self.processor = processor
        self.max_len = max_len

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        file_path = f"{self.audio_dir}/{row[0]}"
        waveform, sr = torchaudio.load(file_path)

        if sr != 16000:
            resampler = torchaudio.transforms.Resample(sr, 16000)
            waveform = resampler(waveform)

        waveform = waveform[:, :self.max_len]
        if waveform.size(1) < self.max_len:
            pad = self.max_len - waveform.size(1)
            waveform = torch.nn.functional.pad(waveform, (0, pad))

        input_values = self.processor(waveform.squeeze().numpy(), sampling_rate=16000, return_tensors="pt").input_values.squeeze(0)
        label = row["label"]
        return input_values, label
```

```python
from torch.utils.data import DataLoader

processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-base")
audio_dir = "/content/asvspoof2017/train/ASVspoof2017_V2_train"

df = train_label
train_dataset = ASVspoofDataset(audio_dir, df, processor)
train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
```

```python
import torch
import torchaudio
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from transformers import Wav2Vec2Model, Wav2Vec2Processor
import pandas as pd
from sklearn.metrics import accuracy_score
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
# STEP 5: Model
class Wav2Vec2FineTune(nn.Module):
    def __init__(self, model_name="facebook/wav2vec2-base"):
        super(Wav2Vec2FineTune, self).__init__()
        self.wav2vec = Wav2Vec2Model.from_pretrained(model_name)
        self.classifier = nn.Sequential(
            nn.Linear(768, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, 1),
            nn.Sigmoid()
        )

    def forward(self, input_values):
        with torch.no_grad():  # freeze feature extractor for stability
            features = self.wav2vec(input_values).last_hidden_state
        pooled = features.mean(dim=1)
        logits = self.classifier(pooled)
        return logits
```

```python
def train(model, dataloader, epochs=3, lr=1e-4):
    model.to(device)
    criterion = nn.BCELoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)

    for epoch in range(epochs):
        model.train()
        total_loss, preds, truths = 0, [], []

        for x, y in dataloader:
            x = x.to(device)
            y = torch.tensor(y, dtype=torch.float32).unsqueeze(1).to(device)

            outputs = model(x)
            loss = criterion(outputs, y)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            preds.extend((outputs > 0.5).int().cpu().numpy())
            truths.extend(y.int().cpu().numpy())

        acc = accuracy_score(truths, preds)
        print(f"Epoch {epoch+1}: Loss = {total_loss:.4f}, Accuracy = {acc:.4f}")
```

```python

model = Wav2Vec2FineTune()
train(model, train_loader, epochs=5)
```
