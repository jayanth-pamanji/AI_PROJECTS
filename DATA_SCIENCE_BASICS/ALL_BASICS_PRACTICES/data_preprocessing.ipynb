```python
headers=["symboling","losses","make","fuel_type","aspiration","num-of-doors","body","wheels","engine-location","wheel-base","length","width","height","weight","engine-type","num-of-cylinders","engine-size","fuel-system","bore","stroke","ratio","horsepower","rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns=headers
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df['symboling']=df['symboling']+1
```

```python
df.head()
```

```python
df["city-mpg"]=235/df["city-mpg"]
```

```python
df.rename(columns={"city-mpg":"city_L/100km"},inplace=True)
```

```python
df.head()
```

```python
df["losses"] = pd.to_numeric(df["losses"], errors='coerce')
mean_value = df["losses"].mean()
df["losses"].fillna(mean_value, inplace=True)
df
```

```python
bins=np.linspace(min(df["price"]),max(df["price"]),4)
```

```python
df.dtypes
```

```python
df["price"] = pd.to_numeric(df["price"], errors='coerce')
mean_value = df["price"].mean()
df["price"].fillna(mean_value, inplace=True)
df
```

```python
bins
```

```python
names=['low','medium','high']
```

```python
df["price_binned"]=pd.cut(df['price'],bins,labels=names,include_lowest=True)
```
